# Q1. To replace '-' with '0'

```python
laliga=laliga.replace(to_replace="-", value= 0)
laliga
```

# Q2. To show Teams with Debut year between 1930 to 1980

```python
a= laliga.loc[:, "Debut"]
debut_yr= list()
for item in a:
    b=int(item[:4])
    debut_yr.append(b)
```

```python
laliga['DebutYear']= debut_yr
```

```python
print("All the teams that played their Debut match in 1930-1980 are:")
for i in range(0,61):
    b= laliga.DebutYear[i]
    c= laliga.Team[i]
    if b>=1930 and b<=1980:
        print(b,c)
```

```python
laliga=laliga.drop("DebutYear", axis=1)
```

# Q3. To show 5 teams with highest points 

```python
laliga['Points']=laliga.Points.astype(int)

sorted_by_points= laliga.sort_values(["Points"], ascending=False)
sorted_by_points.loc[:4,('Points','Team')]
```

# Q4. To define a function Goal_diff_count and get max, min values

```python
laliga['GoalsFor']= laliga.GoalsFor.astype(int)
laliga['GoalsAgainst']= laliga.GoalsAgainst.astype(int)
```

```python
print("The Teams with their respective Goal difference:")
def Goal_diff_count():   
    goal_diff= laliga.GoalsFor- laliga.GoalsAgainst 
    return goal_diff

Goal_diff =Goal_diff_count()

#Defining new DataFrame for Teams and their Goal Difference
n_laliga=pd.DataFrame()
n_laliga["Goal_diff"]= Goal_diff
n_laliga["Teams"]= laliga.Team
n_laliga
```

```python
max_team_index= np.argmax(Goal_diff)
min_team_index= np.argmin(Goal_diff)
print("Max Goal difference:", laliga.Team[max_team_index]) 
print("Min Goal Difference:",laliga.Team[min_team_index])
```

# Q5. Add new Column "Winning Percent"

```python
laliga['GamesWon']= laliga.GamesWon.astype(int)
laliga['GamesPlayed']= laliga.GamesPlayed.astype(int)
laliga["WinningPercent"] = 100* laliga.GamesWon/laliga.GamesPlayed
```

```python
sorted_winperc= laliga.sort_values(["WinningPercent"], ascending=False)
```

```python
print("The top 5 teams with highest Winning Percent are:")
sorted_winperc.loc[:4,('Team','WinningPercent')]
```

# Q6. Group Teams by Best Position and print the sum of points

```python
print("Table showing Team, Points, BestPosition")
laliga.loc[:,('Team','Points','BestPosition')]
```

```python
best_pos= laliga.BestPosition.unique()
```

```python
# To Group Teams by their Best Position

team_bp={}
for i in range(0,61):
    value=laliga.Team[i]
    key=laliga.BestPosition[i]
    
    if key in team_bp:
        team_bp[key].append(value)
    else:
        team_bp[key] = [value]    
# team_bp
```

```python
# To group points and best position of teams

points_bp={}

for i in range(0,61):
    b=laliga.Points[i]
    key=laliga.BestPosition[i]
    
    if key in points_bp:
        points_bp[key].append(b)
    else:
        points_bp[key] = [b] 
    
# points_bp
```

```python
# To add the total points to corresponding Best Position
print("The best position of teams and their total points are:")
for key in points_bp:
    points_list=points_bp[key]
    total_points=sum(points_list)
    bp_dic=team_bp[key]
    if total_points in bp_dic:break
    else:team_bp[key].append(total_points)
    
team_bp
```

```python
#to sum the points of each group

total_pts= list()

for key in points_bp:
    points_list=points_bp[key]
    total_points=sum(points_list)
    total_pts.append(total_points)
    
print(str(total_pts)[0:-1])
```

```python
bestPos=pd.DataFrame()
bestPos["BestPosition"]= best_pos
bestPos["TotalPoints"]= total_pts
sorted_bestPos= bestPos.sort_values(["BestPosition"])
sorted_bestPos.style.hide_index()
```
